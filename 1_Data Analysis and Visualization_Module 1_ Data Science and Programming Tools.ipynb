Q1.a Importing numpy library and pyplot from matplotlib library

```python
#Importing required libraries in one cell followed as a best practise in every Jupyter notebook

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Q2.b. Plot sin and cosine waves from –π to +π

```python
#linspace and inbuilt function in numpy 
#used to create an evenly spaced sequence in a specified interval.
val = np.linspace(-np.pi,np.pi,100)

#defining the values and inputs for sin and cos 
plt.plot(val, np.sin(val),  label = "sin")
plt.plot(val, np.cos(val), label = "cos")


#Displaying the necessary info on graph to make it better 
plt.title("sin and cosin from –π to +π")
plt.xlabel("Degree")
plt.ylabel("sin and cos values")

plt.legend() #Clarifies the info on graph

#Displaying the graph
plt.show()
```
