```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

```

```python

# Load the SMS Spam Collection Dataset (you need to provide the file path)
df = pd.read_csv(r'D:\1_AJINKYA_PAWAR\3_SOFTWARE JOB\1_CDAC\3_AI_Course\Lab_Assignment_Module_3\3_3_NLP_2Hrs\SMS-spam-collection-dataset.csv', sep='\t', names=['label', 'message'], encoding='ISO-8859-1')


# Ensure the 'message' column is of string data type
df['message'] = df['message'].astype(str)

# Data Cleaning
df['message'] = df['message'].str.lower()  # Convert to lowercase
df['message'] = df['message'].str.replace('[^a-zA-Z\s]', '', regex=True)  # Remove special characters and numbers

# TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
X = tfidf_vectorizer.fit_transform(df['message'])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, df['label'], test_size=0.2, random_state=42)

# Train a Multinomial Naive Bayes classifier
clf = MultinomialNB()
clf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
confusion = confusion_matrix(y_test, y_pred)
classification_report_str = classification_report(y_test, y_pred)

# Print the results
print(f"Accuracy: {accuracy}")
print("Confusion Matrix:")
print(confusion)
print("Classification Report:")
print(classification_report_str)
```
