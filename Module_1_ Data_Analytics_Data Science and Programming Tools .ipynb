#### Q5: 5. Find the items which have MRP higher than 200 and count how many of them are low fat / LF, reg fat / reg


```python
# items which have MRP higher than 200

MRP_outlets = df.loc[ df['Item_MRP'] >= 200] 

MRP_outlets
```

```python
# Items whose MRP is greater than 200 and have low fat content 
lowfat_outlets = MRP_outlets.loc[MRP_outlets['Item_Fat_Content'] == 'Low Fat'] 

lowfat_outlets.count()

#Answer: Total 862 items have MRP greater than 200 and low fat content 
```

```python
# Items whose MRP is greater than 200 and have reg fat content 
regfat_outlets = MRP_outlets.loc[MRP_outlets['Item_Fat_Content'] == 'Regular'] 

regfat_outlets.count()

#Answer: Total 488 items have MRP greater than 200 and low fat content 
```
