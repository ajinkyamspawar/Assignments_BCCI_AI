#### Q1. Use -wheather_data_cities.csv draw a bar graph and pie chart on x axis use city names and on y axis use max of temperature

```python
#Importing required libraries in one cell followed as a best practise in every Jupyter notebook

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing dataframe 

df = pd.read_csv(r"D:\1_AJINKYA_PAWAR\3_SOFTWARE JOB\1_CDAC\3_AI_Course\Lab_Assignment_Module_1\weather_data_cities.csv")

df

#Dataframe imported and displayed as below 
```

```python
# Grouping the data by city and find the maximum temperature for each city

max_temp = df.groupby('city')['temperature'].max().reset_index()

max_temp
```

#### Q1: A. Records of every item:: describe various statistics about every item type record

```python
# Now we will draw a bar graph

plt.bar(max_temp['city'], max_temp['temperature'])
plt.xlabel('City')
plt.ylabel('Maximum Temperature')
plt.title('Maximum Temperature by City')
plt.show()
```

```python
# Now we will draw a pie chart

plt.pie(max_temp['temperature'], labels=max_temp['city'] , autopct='%1.1f%%')
plt.title('Percentage of Maximum Temperature by City')
plt.show()
```

## 2. Use company_sales_data.csv files and analyze the dataset and plot graphs accordingly

```python
df = pd.read_csv(r"D:\1_AJINKYA_PAWAR\3_SOFTWARE JOB\1_CDAC\3_AI_Course\Lab_Assignment_Module_1\company_sales_data.csv")
df
```

#### 2.a: plot profit for all months use line graph.


```python
plt.plot(df['month_number'], df['total_profit'], marker='o')
plt.xlabel('Month')
plt.ylabel('Profit')
plt.title('Profit for All Months')
plt.show()
```

#### 2.b: plot sales data for all product using pie chart graph


```python
#Calculate the total sales for each product


product_sales = df.iloc[:, 1:7].sum()
product_sales
```

```python
# Plotting a pie chart

plt.pie(product_sales, labels=product_sales.index, autopct='%1.1f%%')
plt.title('Sales Data for All Products')
plt.show()
```

#### 2.c: plot face cream sales data using scatter plot


```python
plt.scatter(df['month_number'], df['facecream'])
plt.xlabel('Month')
plt.ylabel('Face Cream Sales')
plt.title('Face Cream Sales Data')
plt.show()
```

#### 2.d: compare face cream and toothpaste using bar graph

```python
months = df['month_number']
facecream_sales = df['facecream']
toothpaste_sales = df['toothpaste']

plt.bar(months - 0.2, facecream_sales, width=0.4, align='center', label='Face Cream')
plt.bar(months + 0.2, toothpaste_sales, width=0.4, align='center', label='Toothpaste')

plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Comparison of Face Cream and Toothpaste Sales')
plt.legend()
plt.show()
```
