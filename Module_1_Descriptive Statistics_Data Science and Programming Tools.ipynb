```python
#Importing required libraries in one cell followed as a best practise in every Jupyter notebook

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\1_AJINKYA_PAWAR\3_SOFTWARE JOB\1_CDAC\3_AI_Course\Lab_Assignment_Module_1\house_prices.csv")

df
```

####
Consider the dataset house_prices describe 35 metrics about a House including its final Sale Price.
1. The task is to perform exploratory data analysis on the dataset and generate Descriptive statistics such as Mean, Median, Mode, Variance, Skew, Kurtosis, Correlation and Quantiles of various features present in the data.

```python
# Calculating descriptive statistics
mean = df.mean()
median = df.median()
mode = df.mode().iloc[0]  # Mode can have multiple values, so we'll take the first one
variance = df.var()
skewness = df.skew()
kurtosis = df.kurt()
quantiles = df.quantile([0.25, 0.5, 0.75])

# Calculating the  correlation, we will use df.corr()
correlation = df.corr()

# Printing  the results
print("Descriptive Statistics:")
print("Mean:")
print(mean)
print("\nMedian:")
print(median)
print("\nMode:")
print(mode)
print("\nVariance:")
print(variance)
print("\nSkewness:")
print(skewness)
print("\nKurtosis:")
print(kurtosis)
print("\nQuantiles:")
print(quantiles)
print("\nCorrelation:")
print(correlation)
```
