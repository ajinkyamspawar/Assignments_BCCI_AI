#### Q1. Perform basic data cleaning.

#### A. Apply at least 2 feature selection algorithms.





#### 1.  Correlation-based Feature Selection: 

```python
#Perforimng necessary data cleaning and then applying feature selection algorithms

# Removing unnecessary columns
columns_to_drop = ['timestamp', 'location', 'tag', 'gender', 'otherdetails', 'cityid', 'dmaid', 'rowNumber', 'Masters_Degree', 'Bachelors_Degree', 'Doctorate_Degree', 'Highschool', 'Some_College', 'Race_Asian', 'Race_White', 'Race_Two_Or_More', 'Race_Black', 'Race_Hispanic', 'Race', 'Education']
data = df.drop(columns=columns_to_drop)

# Replacing 'NA' with NaN
df = df.replace('NA', pd.NA)

# Till here basic data clearning is done. Now we will move ahead with following questions. 
# Now, we have a cleaned dataframe without any missing values.


# Converting numeric columns to numeric type
numeric_columns = ['totalyearlycompensation', 'yearsofexperience', 'yearsatcompany', 'basesalary', 'stockgrantvalue', 'bonus']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Filtering numeric columns for correlation calculation
numeric_data = df.select_dtypes(include='number')

# Performing correlation-based feature selection
correlation_matrix = numeric_data.corr()
correlation_threshold = 0.5
correlated_features = correlation_matrix['totalyearlycompensation'].abs() > correlation_threshold
selected_features_correlation = correlation_matrix.columns[correlated_features].tolist()

print(selected_features_correlation)
```

#### 2: Variance Threshold based feature selection:

```python

# Removing unnecessary columns
columns_to_drop = ['timestamp', 'location', 'tag', 'gender', 'otherdetails', 'cityid', 'dmaid', 'rowNumber', 'Masters_Degree', 'Bachelors_Degree', 'Doctorate_Degree', 'Highschool', 'Some_College', 'Race_Asian', 'Race_White', 'Race_Two_Or_More', 'Race_Black', 'Race_Hispanic', 'Race', 'Education']
data = df.drop(columns=columns_to_drop)

# Replacing 'NA' with NaN
data = data.replace('NA', pd.NA)

# Converting numeric columns to numeric type
numeric_columns = ['totalyearlycompensation', 'yearsofexperience', 'yearsatcompany', 'basesalary', 'stockgrantvalue', 'bonus']
data[numeric_columns] = data[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Filtering numeric columns for Variance Threshold feature selection
numeric_data = data.select_dtypes(include='number')

# Performing Variance Threshold feature selection
variance_threshold = 0.1
selector = VarianceThreshold(threshold=variance_threshold)
selector.fit(numeric_data)
selected_features_variance = numeric_data.columns[selector.get_support()].tolist()

print(selected_features_variance)
```

#### B. Applying linear regression.  and C. Visualising the result 



```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Using selected features from either filter
selected_features = selected_features_correlation  # You can choose filter features here

# Splitting the data into features and target again, this time using selected features
X_selected = df[['basesalary' ]]
y = df[['totalyearlycompensation']]

# Performing Linear Regression
regressor = LinearRegression()
regressor.fit(X_selected, y)

# Visualizing the results
plt.scatter(X_selected, y, color='blue', label='Actual Data')
plt.plot(X_selected, regressor.predict(X_selected), color='red', label='Linear Regression')
plt.xlabel('Selected Features')
plt.ylabel('totalyearlycompensation')
plt.legend()
plt.title('Linear Regression Results')
plt.show()
```

Hypothesis : 

Null Hypothesis (H0):
There is no  difference in the performance of the linear regression model using 
the 'basesalary' feature and the linear regression model using the 'bonus' feature in predicting the 
'totalyearlycompensation' target variable.

Alternative Hypothesis (H1):
There is a difference in the performance of the linear regression model using the
'basesalary' feature and the linear regression model using the 'bonus' feature in predicting the
'totalyearlycompensation' target variable.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from scipy.stats import ttest_rel

# Using selected features from either filter  methods
selected_features = selected_features_correlation  

# Split the data into features and target again, this time using selected features
X_selected_basalary = df[['bonus']]
y = df['totalyearlycompensation']

# Performing Linear Regression
regressor = LinearRegression()

# Cross-validation
num_folds = 5
mse_scores_basalary = cross_val_score(regressor, X_selected_basalary, y, scoring='neg_mean_squared_error', cv=num_folds)
rmse_scores_basalary = (-mse_scores_basalary) ** 0.5  # Convert negative MSE scores to RMSE scores

# Printing the RMSE scores for each fold
print("RMSE Scores for 'basesalary' feature:", rmse_scores_basalary)

# Calculating the mean and standard deviation of RMSE scores
mean_rmse_basalary = rmse_scores_basalary.mean()
std_rmse_basalary = rmse_scores_basalary.std()

print("Mean RMSE for 'basesalary' feature:", mean_rmse_basalary)
print("Standard Deviation RMSE for 'basesalary' feature:", std_rmse_basalary)

# Now, repeat the process for the second feature 'bonus'
X_selected_bonus = df[['basesalary']]
y = df['totalyearlycompensation']

# Performing Linear Regression
regressor = LinearRegression()

# Cross-validation
mse_scores_bonus = cross_val_score(regressor, X_selected_bonus, y, scoring='neg_mean_squared_error', cv=num_folds)
rmse_scores_bonus = (-mse_scores_bonus) ** 0.5  # Convert negative MSE scores to RMSE scores

# Printing the RMSE scores for each fold
print("RMSE Scores for 'bonus' feature:", rmse_scores_bonus)

# Calculating the mean and standard deviation of RMSE scores
mean_rmse_bonus = rmse_scores_bonus.mean()
std_rmse_bonus = rmse_scores_bonus.std()

print("Mean RMSE for 'bonus' feature:", mean_rmse_bonus)
print("Standard Deviation RMSE for 'bonus' feature:", std_rmse_bonus)

# Performing t-test to compare the two models
t_stat, p_value = ttest_rel(rmse_scores_basalary, rmse_scores_bonus)
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

#### E. Write the conclusion of the t-test.

As the p-value is 0.00015128398588537043, it is clear that it lies in the significant region. 

Hence, we reject the null hypothesis. There is a difference in the performance of the linear regression 
model using the 'basesalary' feature and the linear regression model using the 'bonus' feature in predicting 
the 'totalyearlycompensation' target variable.
