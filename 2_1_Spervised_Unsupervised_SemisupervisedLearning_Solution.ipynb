1. Unsupervised Learning problem using K-means Clustering on Country-data.csv dataset.
Perform K-means Clustering taking K=3 and plot graph for child mortality vs countries. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

```

```python
# Loading the dataset
data = pd.read_csv(r"D:\1_AJINKYA_PAWAR\3_SOFTWARE JOB\1_CDAC\3_AI_Course\Lab_Assignment_Module_2\2_1_Spervised_Unsupervised_SemisupervisedLearning_30Min\Country-data.csv")

# Displaying the first few rows of the dataset to understand its structure
print(data.head())
```

```python
# Selecting the relevant features for clustering
X = data[['child_mort', 'country']]

# Standardizing the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X[['child_mort']])

# Applying K-means clustering with K=3
kmeans = KMeans(n_clusters=3, random_state=42)
X['Cluster'] = kmeans.fit_predict(X_scaled)

# Visualizing the clusters
plt.figure(figsize=(12, 6))
for i in range(3):
    cluster = X[X['Cluster'] == i]
    plt.scatter(cluster['country'], cluster['child_mort'], label=f'Cluster {i}')

plt.xlabel('country')
plt.ylabel('child mortality')
plt.title('K-means Clustering (K=3) - child mortality vs countries')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
